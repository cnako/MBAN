## Purpose
The Massachusetts state department of transportation recognized that even for relatively small trips automobiles are being used throughout the state. The department has decided to encourage the use of bicycles and make cities more bicycle friendly. Keeping in mind the increasing population and lack of space in Boston the state wants to encourage people to become everyday cyclists in the Boston. They also want to promote health and wellbeing and curb greenhouse gas emissions from vehicles. Boston has a bikesharing service called BlueBikes which is becoming an integral part of the cityâ€™s public transit network. We want to help the state rethink bicycle mobility in Boston by proposing improvements such that the bikeshare network can benefit from the business distribution in the city. We would do that by analyzing the following questions.

Questions
1. Which type of business correlates with a higher usage of bike sharing services?
2. Does this change if we look at the different user types (subscription vs. one-off riders)?

## Methodology
In order to answer the questions above, we mainly utilize four datasets. The first dataset is Bluebikes Trip Data, which contains user trip data from June - August 2021. The next datasets are the Foursquare-Venues and Foursquare-Categories dataset, to analyze business in Boston with respect to their latitude, longitude, postal code, type of business, categories and sub-categories. Lastly, we use the zipcode dataset to map locations and zones across the datasets - Foursquare and Bluebikes. However, the datasets have some limitations so that we changed have chosen to answer the first two questions of our original proposal. Further we plan to carry out the following tasks in orde to answer our business questions.
